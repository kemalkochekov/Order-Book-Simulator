```python
# Retrieve stock data
import yfinance as yf
import pandas as pd
ticker = "RUB=X"

dataF = yf.download(ticker, start="2023-04-10", end="2023-06-01", interval='15m')
dataF.iloc[:,:]
#dataF.Open.iloc
```

```python
def signal_generator(df):
    open = df.Open.iloc[-1]
    close = df.Close.iloc[-1]
    previous_open = df.Open.iloc[-2]
    previous_close = df.Close.iloc[-2]
    
    # Bearish Pattern & Selling
    if (open>close and 
    previous_open<previous_close and 
    close<previous_open and
    open>=previous_close):
        return 1

    # Bullish Pattern & Buyying
    elif (open<close and 
        previous_open>previous_close and 
        close>previous_open and
        open<=previous_close):
        return 2
    
    # No clear pattern
    else:
        return 0

signal = []
signal.append(0)
for i in range(1,len(dataF)):
    df = dataF[i-1:i+1]
    signal.append(signal_generator(df))
#signal_generator(data)
dataF["signal"] = signal
```

```python
dataF.signal.value_counts()
#dataF.iloc[:, :]
```

```python
# Buy Price datas
Buy_Price=[]
for signal, buy_price in zip(dataF["signal"], dataF["Open"]):
    if signal == 2:
        Buy_Price.append(buy_price)
# Save Buy_Price to file
with open("stock_data_buy.txt", "w") as file:
    for price in Buy_Price:
        file.write(str(price) + "\n")
```
